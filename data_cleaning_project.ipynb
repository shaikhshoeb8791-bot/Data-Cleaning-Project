```python
import pandas as pd

# Load data
df = pd.read_excel("Dataset for Data Analytics 02 (1).xlsx")

# Check data
print("First 5 rows:")
display(df.head())

print("Dataset shape:")
print(df.shape)

# Data information
print("\nData Info:")
df.info()

# Missing values
print("\nMissing Values:")
print(df.isnull().sum())

# Remove duplicates
print("\nDuplicate Rows:")
print(df.duplicated().sum())

df = df.drop_duplicates()

# Date format correction
if 'Date' in df.columns:
    df['Date'] = pd.to_datetime(df['Date'])

# Clean text columns
df = df.apply(lambda x: x.str.strip() if x.dtype=="object" else x)

# Save cleaned file
df.to_excel("Cleaned_Data.xlsx", index=False)

print("\nCleaning Completed!")
```
