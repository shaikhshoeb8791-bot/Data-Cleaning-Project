```python
import pandas as pd

df = pd.read_excel("Dataset for Data Analytics (2).xlsx")

df.head()
```

```python
df.columns
```

```python
import sqlite3

conn = sqlite3.connect("project.db")
```

```python
df.to_sql("data", conn, if_exists="replace", index=False)
```

```python
query = "SELECT * FROM data"
pd.read_sql(query, conn).head()
```

```python
query = """
SELECT * FROM data
WHERE df.columns[0] IS NOT NULL
"""
pd.read_sql(query, conn).head()
```

```python
query = "SELECT * FROM data"
pd.read_sql(query, conn).head()
```

```python
query = """
SELECT * FROM data
WHERE Quantity > 2
"""
pd.read_sql(query, conn)
```

```python
query = """
SELECT * FROM data
ORDER BY TotalPrice DESC
"""
pd.read_sql(query, conn).head()
```

```python
query = """
SELECT Product, SUM(Quantity) AS Total_Quantity
FROM data
GROUP BY Product
ORDER BY Total_Quantity DESC
"""
pd.read_sql(query, conn)
```

```python
query = """
SELECT SUM(TotalPrice) AS Total_Revenue
FROM data
"""
pd.read_sql(query, conn)
```

```python
query = """
SELECT AVG(TotalPrice) AS Avg_Order_Value
FROM data
"""
pd.read_sql(query, conn)
```

```python
query = """
SELECT Product, SUM(TotalPrice) AS Revenue
FROM data
GROUP BY Product
ORDER BY Revenue DESC
LIMIT 5
"""
pd.read_sql(query, conn)
```

```python
query = """
SELECT PaymentMethod, COUNT(*) AS Orders
FROM data
GROUP BY PaymentMethod
ORDER BY Orders DESC
"""
pd.read_sql(query, conn)
```
