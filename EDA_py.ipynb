```python
import pandas as pd

df = pd.read_csv("America.csv", sep="\t")

df.head()
```

```python
print(df.columns)
```

```python
df["Product"].value_counts()
```

```python
df["OrderStatus"].value_counts()
```

```python
import matplotlib.pyplot as plt
```

```python
df["OrderStatus"].value_counts().plot(kind="bar", figsize=(6,4))

plt.title("Order Status")
plt.xlabel("Status")
plt.ylabel("Count")
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

plt.hist(df["Quantity"], bins=10, edgecolor="black")

plt.title("Distribution of Quantity")
plt.xlabel("Quantity")
plt.ylabel("Frequency")

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.hist(df["TotalPrice"], bins=10, edgecolor="black")

plt.title("Distribution of Total Price")
plt.xlabel("Total Price")
plt.ylabel("Frequency")

plt.show()
```

```python
plt.figure(figsize=(6,5))

plt.boxplot(df["TotalPrice"])

plt.title("Box Plot of Total Price")

plt.show()
```

## Key Observations

1. The dataset contains customer orders with details such as product, quantity, payment method, and total price.

2. Some products received more orders than others, indicating higher customer demand.

3. Most orders were concentrated in one or two order status categories (such as Delivered or Shipped).

4. The quantity distribution shows that most customers purchased a small number of items, while larger quantities were less common.

5. The total price distribution indicates that most orders fall within a certain price range, with a few high-value orders acting as outliers.

6. The box plot suggests the presence of a few outliers in the Total Price column, representing unusually expensive orders.

### Conclusion

The exploratory data analysis helped identify product demand, order patterns, and price distribution. Overall, the dataset provides useful insights into customer purchasing behavior and order trends.
